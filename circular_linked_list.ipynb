```python
# Circular Linked List Code
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next

class CLL:
    def __init__(self,end=None):
        self.end = end
    
    def is_empty(self):
        return self.end == None
    
    def list_len(self):
            if self.end == None:
                print(0)
            else:
                temp = self.end.next
                count = 0
                while True:
                    count +=1
                    temp = temp.next
                    if temp == self.end.next:
                        break
                print(count)
    
    def insert_at_first(self,val):
        new_node = Node(val)
        if self.end == None:
            new_node.next = new_node
            self.end = new_node
        else:
            new_node.next = self.end.next
            self.end.next = new_node   

    def insert_at_last(self,val):
        new_node = Node(val)
        if self.end != None:
            new_node.next = self.end.next
            self.end.next = new_node
            self.end = new_node
        else:
            new_node.next = new_node
            self.end = new_node

    def insert_after(self,node_val,val):
        new_node=Node(val)
        if self.end != None:
            temp = self.end.next
            if temp.data == node_val:
                new_node.next = temp.next
                temp.next = new_node
            else:
                temp = self.end.next.next
                while temp != self.end.next:
                    if temp.data == node_val and temp!= self.end:
                        new_node.next = temp.next
                        temp.next = new_node
                        break
                    elif temp.data == node_val and temp == self.end:
                        new_node.next = self.end.next
                        self.end.next = new_node
                        self.end = new_node
                        break
                    else:
                        temp = temp.next
        else:
            print("Cannot insert node: List is Empty !!!")

    def delete_first(self):
        if self.end == None:
            print('List is empty..')
        elif self.end.next == self.end:
            self.end = None
        elif self.end!= None:
            temp = self.end.next
            self.end.next = self.end.next.next
            temp.next = None
            
    def delete_last(self):
        if self.end == None:
            print("List is empty...")
        elif self.end.next == self.end:
            self.end = None
        else:
            temp = self.end.next
            while temp.next!= self.end:
                temp = temp.next
            temp.next = self.end.next
            self.end = temp
        
    def print_list(self):
        if self.end != None:
            temp = self.end.next
            while True:
                print(temp.data,end=" ")
                temp = temp.next
                if temp == self.end.next:
                    break
        print()
    
    def __iter__(self):
        if self.end == None:
            return CLL_Iterator(None)
        else:
            return CLL_Iterator(self.end.next)

class CLL_Iterator:
    def __init__(self,start):
        self.current = start
        self.start = start
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.current == None:
            raise StopIteration
        if self.current == self.start and self.count == 1:
            raise StopIteration
        else:
            self.count = 1
        data = self.current.data
        self.current = self.current.next
        return data
        
            
```
